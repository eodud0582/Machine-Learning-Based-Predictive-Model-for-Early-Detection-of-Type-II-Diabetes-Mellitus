##### 혈압 혈당 데이터
- BMI: 체질량지수(Body Mass Index)
    - 몸무게(kg)를 키의 제곱(㎡)으로 나눈값
    - 비만인 사람들이 아닌 사람들에 비해 당뇨병 발생위험이 6배
    - 실제로 우리나라 당뇨병 환자의 74.9%가 과제충이거나 비만

```
    한국인의 체질량 지수에 따른 비만 판정 기준
    - <18.5: 저체중
    - 18.5-22.9: 정상
    - 23.0-24.9: 비만 전단계(과체중)
    - 25.0-29.9: 1단계 비만
    - 30.0-34.9: 2단계 비만
    - 35.0 이상: 3단계 비만

    (참고) BMI Bins 
    - <19: Starvation
    - 19–24: Normal
    - 25–30: Overweight
    - 31–40: Obese
    - 40: Very Obese
```

- SBP: 수축기혈압(Systolic Blood Pressure); 심장이 수축할 때 혈관에 가해지는 압력
- DBP: 이완기혈압(Diastolic Blood Pressure); 심장이 이완할 때 혈압에 가해지는 압력



```
    (참고) DBP Bins 
    - <61: Very low
    - 61–75: Low
    - 75–90: Normal
    - 91–100: High
    - 100: Hypertension
```

```
- 한 통계에 따르면 일반인에서는 약 30%가 고혈압을 갖고 있는 반면, 당뇨병 환자에서는 65% 이상이 고혈압을 갖고 있다. 우리나라 당뇨병 유병자 중 30세 이상 55.3%, 65세 이상 71.2%가 고혈압을 동반한다.
- 정상혈압은 수축기혈압 <120mmHg, 이완기혈압 <80mmHg 아래여야 한다.
- 고혈압(수축기혈압≥140 또는 이완기혈압≥90 mmHg 진단 기준)은 제 1 형 당뇨병 환자의 약 25%, 제 2 형 당뇨병 환자의 약 50%에서 이환되어 있다.
- 수축기 혈압 ≥ 160 mmHg 또는 이완기 혈압 ≥ 100 mmHg인 경우는 즉각적인 약물치료 시작이 필요합니다.
- 현재 우리나라 당뇨병학회에서 권고하는 일반적인 당뇨병 환자의 목표 혈압은 140/85 mmHg 미만입니다. 최근 미국심장학회에서 고혈압 진단 기준을 140/90 mmHg에서 130/80 mm Hg로 낮췄으며 당뇨병 환자에게서도 130/80 mmHg 이상부터 약물치료를 고려하도록 권고하고 있습니다.
- 당뇨병 환자가 고혈압일 가능성은 당뇨병이 아닌 사람에 비해 최고 2배 정도 높으므로 당뇨병 합병증을 예방하기 위해서는 혈당 조절 외에도 혈압과 콜레스테롤관리에 신경을 써야 한다.
```

- PR: 맥박수(Pulse Rate)
    - 1분에 60-100회: 정상 맥박
    - 60회 미만: 서맥(서맥성 부정맥)
    - 100회 이상: 빈맥(빈맥성 부정맥)
    - 맥박수가 분당 90회 이상인 사람의 당뇨병 위험도는 60회 미만인 사람보다 남성이 3.85배, 여성이 3.34배 높았다
- HbA1c: 당화혈색소(HbA1c)
    - 검사의 목적: 1) 당뇨 및 당뇨병전기를 선별하고 진단하기 위함, 2) 환자의 당뇨를 모니터하고 치료 결정을 내리는데 도움을 주기 위함 (최초로 당뇨로 진단받았을 때 그리고 일년에 2-4회 검사; 건강검진의 일부로서 또는 당뇨의 증이 있을 경우)
    - 해석
        - ≤ 5.6%: 참고치 이내
        - 5.7-6.4%: 당뇨 전단계
        - ≥6.5%: 당뇨
- FBG: 공복시 혈당(Fasting Plasma Glucose)
    - 혈중 포도당 농도를 혈당이라고 한다. 원래 혈액에는 일정 수준의 포도당이 내재해 있다. 그러나 일정 수준을 넘어서면 당이 소변으로 넘쳐 나와 당뇨병이 된다.
    - 최소한 8시간 동안 공복 후 측정한 혈당이 126mg/dl 이상인 경우 당뇨 진단
    - <70: 저혈당
    - 70-99 mg/dL: 정상 공복 혈당
    - 100-125: 공복혈당장애(당뇨병 전단계)
    - 126<=: 당뇨병
    
    (https://labtestsonline.kr/tests/glucose)

##### 콜레스테롤 관련
- TC: 총콜레스테롤(Total Cholesterol); 
    - 총콜레스테롤 = HDL+LDL+(TG/5)
    - 당뇨병 유병자 중 72%가 고콜레스테롤혈증을 동반하고 있어 질환 관리가 필요하다
    - <200 mg/dL: 바람직함(desirable)
    - 200-239: 경계성으로 높음(borderline high)
    - 240<=: 높은 위험도/이상지질혈증 진단(high)
        - 수치가 200~239㎎/㎗여도 LDL콜레스테롤과 중성지방이 많다면 이상지질혈증으로 진단
     
      (https://labtestsonline.kr/tests/cholesterol)
- TG: 중성지방(Triglyceride)
    - 음식물을 통해 섭취한 칼로리가 충분히 소비되지 못하고 간(肝)에서 합성되거나 장(腸)에서 흡수돼 생기는 지방이다.
    - <150: 정상
    - 150-199: 경계치
    - 200-499: 높음
    - 500<=: 매우 높음
    - 40-155: 정상 / 50-200이 정상이라는 곳도 있다
- LDL: 저밀도 콜레스테롤(나쁜 콜레스테롤; Low-density Lipoprotein Cholesterol)
    - 일명 '나쁜 콜레스테롤'이라고 불리는데, 체내에 LDL콜레스테롤이 많으면 혈관벽에 쌓이면서 혈액이 끈적해지고, 혈전이 잘 생긴다. 혈전이 생기면 혈관이 좁아지고, 심한 경우 혈관이 막혀 심근경색이나 뇌졸중 등이 생긴다.
    - <100: 정상(optimal)
    - 100-129: 정상 이상(near optimal/above optimal)
    - 130-159: 약간 높음(borderline high)
    - 160-189: 높음(high)
    - 190<=: 아주 높음(very high)
    - 5-140: 정상
    - 일반적인 건강한 성인의 경우 LDL 콜레스테롤을 160이하를 정상
    - 고혈압, 당뇨 등의 질환이 있는 경우, 나이가 많은 경우, 흡연자, 심혈관질환 가족력이 있는 환자는 130이하
    - 당뇨병이 있거나 경동맥 초음파에서 의미 있는 협착을 진단받는 경우는 100이하
- HDL: 고밀도 콜레스테롤(좋은 콜레스테롤; High-density Lipoprotein Cholesterol)
    - 일명 '좋은 콜레스테롤'이라고 불리는데, 혈액 속 나쁜 콜레스테롤(LDL)을 체외로 배출시키는 역할을 한다.
    - <40: 낮음(low)
    - 40-60: 보통
    - 60<=: 높음(high)
    - 32-70: 정상 / 40이하면 문제 라는 곳도 있다

##### 간 및 신장 관련
- Alb: 알부민(Alb); 혈액의 액상 부분인 혈장에 가장 많이 존재하는 단백질로서 혈관으로부터 액체의 유출을 막고, 조직에 영양분을 제공하며, 호로몬, 비타민 약물, 칼슘 같은 이온을 신체로 전달한다.
    - 알부민은 간에서 만들어지며, 알부민의 농도는 간이 손상되거나, 신장질환이 있거나, 영양실조이거나, 염증 및 쇼크에 빠진 경우 감소하는 경향을 보인다.
    - 따라서, 다른 간 검사 결과와는 달리 알부민은 수치가 낮은 것이 문제가 되는 경우가 많다; 수치가 높게 나오는 이유는 많은 경우 탈수증이 원인이다.
    - 3.5-5.2g/dL: 정상
- BUN: 혈중요소질소(Blood Urea Nitrogen); 요소는 간에서 단백이 분해될 때 생기는 노폐물로, 혈액을 통해 신장에 도달하여 소변으로 배출됩니다. 몸 안에서 생산된 요소의 대부분은 신장에서 제거되므로 혈액 속의 요소질소 농도를 확인함으로써 신장이 얼마나 잘 작동하는지 확인할 수 있습니다.
    - 신장기능이 안 좋으면 배설을 못해 증가한다
    - 높아진 BUN 농도는 신장 기능의 저하를 의미(급성 혹은 만성 신장질환을 뜻함); 낮은 BUN 농도는 심한 간질환, 정상 임신, 영양실조에서 나타날 수 있다
    - 6-20mg/dL: 정상 (https://m.amc.seoul.kr/asan/mobile/healthinfo/management/managementDetail.do?managementId=102)
- Cr: 크레아티닌(Creatinine); 크레아티닌은 근육에서 생산되는 노폐물로, 대부분 신장에서 제거되므로 혈액 안의 크레아티닌 농도는 신장 기능을 보여주는 지표가 됩니다. 생산되는 농도는 근육의 양에 따라 조금씩 달라질 수 있으며, 혈액 내의 크레아티닌 농도를 확인합니다.
    - 혈액 내 크레아티닌의 증가는 신장 내 혈관 손상, 신장의 세균 감염, 세뇨관 손상, 신장 결석, 요관 폐색, 신장으로의 혈류 감소, 전립선 질환 등을 의미
    - 신장기능이 안 좋으면 배설을 못해 증가한다
    - 크레아티닌은 근육이 분해될 때 나오는 노폐물로, 근육의 양이 많은 경우에는 일반적인 수치보다 높을 수 있으므로 신장과 체중을 고려하여 검사 결과를 해석하기도 합니다. 일반적으로 임신 기간에는 약간 낮은 상태로 유지됩니다.
    - 0.5-1.4mg/dL: 정상 (https://amc.seoul.kr/asan/healthinfo/management/managementDetail.do?managementId=104)
    - 남성 : 0.6 <= Cr < 1.2 정상
    - 여성 : 0.5 <=

- CrCl: 크레아티닌 청소율(Creatinine Clearance); 신장에서 제거되는 크레아티닌의 양을 이용하여 사구체여과율(GFR:Glomerular filtration rate)을 추정하는 방법입니다. 사구체가 손상되거나 신장으로의 혈액 순환이 감소되면 제거되는 크레아티닌이 적고 사구체가 분당 여과할 수 있는 능력 또한 감소됩니다. 이를 토대로 24시간 동안 만들어진 소변을 통해 여과된 크레아티닌을 확인하여 신장 내 사구체여과율을 확인합니다.

- BUN/Cr ratio: 신장 질환의 여부 및 문제가 되는 곳을 찾아낼 때 BUN/Cr 비율을 이용

    - 10 < BUN/Cr < 20: 정상 범위; 그러나 BUN과 Cr이 정상보다 높은 수치; 신부전, 사구체질환, 뇨세관 질환 등의 가능성
    - BUN/Cr < 10: 신장 자체에 이상이 있어서 BUN의 재흡수가 잘 일어나지 않는 상태

    - AST/ALT(AST 및 ALT 비율) = 간 기능의 데미지를 파악하는데 효과적인 지표로, 일반적으로 평균값을 벗어나는 불안정한 AST와 ALT의 데이터로부터 AST/ALT 값이 1보다 작다면 급성의 간손상을 그리고 1보다 크게된다면 지속적인 간의 손상을 감안해볼 수 있다.
